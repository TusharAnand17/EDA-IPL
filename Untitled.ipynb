# Most win in IPL

```python
df = df.replace(['Rising Pune Supergiants'],'Rising Pune Supergiant')
```

```python
mostWin = df['winner'].value_counts()
mostWin = pd.DataFrame(mostWin)
mostWin
```

```python
data = np.array(mostWin['winner'])
label = [i for i in mostWin.index]
plt.figure(figsize = (10,12))
plt.pie(data,labels = label,autopct = '%1.1f%%');
```

# Most winner in eliminator

```python
eliminatorwinner = df[df['eliminator']=='Y']
eliminatorwinner = eliminatorwinner[['winner','eliminator']]
sns.countplot(x = 'winner',data = eliminatorwinner);
plt.xticks(rotation= 30,ha= 'right');
plt.xlabel('Teams');
plt.ylabel('Most No. of Wins in Eliminator');
```

# Toss decision

```python
tossdecisionteam = df.groupby(df['toss_winner'],as_index=False)[['toss_decision']].count()
tossdecisionteam

batdecision = df[df['toss_decision']=='bat']
batdecision = batdecision[['toss_decision','toss_winner']]
batdecision=batdecision.groupby('toss_winner',as_index=False)[['toss_decision']].count()
batdecision.rename(columns={'toss_decision':'bat'},inplace=True)

fielddecision = df[df['toss_decision']=='field']
fileddecision = fielddecision[['toss_decision','toss_winner']]
fielddecision=fielddecision.groupby('toss_winner',as_index=False)[['toss_decision']].count()
fielddecision.rename(columns={'toss_decision':'field'},inplace=True)
```

```python
tossdecisionteam['toss_decision']
```

```python
batdecision
```

```python
fielddecision
```

```python
final = pd.merge(tossdecisionteam,batdecision,on="toss_winner")
Fullfinal = pd.merge(final,fielddecision,on='toss_winner')
Fullfinal
```

```python
plt.plot(Fullfinal['toss_winner'],Fullfinal['bat'])
plt.plot(Fullfinal['toss_winner'],Fullfinal['field'])
plt.legend(['bat','field']);
plt.xticks(rotation = 90);
```

```python
famousVenuename = df['venue'].value_counts().index[0:10]
famousVenuecount = df['venue'].value_counts().values[0:10]
famousvenue = sns.barplot(x=famousVenuename,y = famousVenuecount)
plt.xticks(rotation = 90,ha='right');

for bars in famousvenue.containers:
    famousvenue.bar_label(bars)
    
plt.xlabel('Venue')
plt.ylabel('Match Played')
plt.title('Most Famous Venue');
```
